3. Qual a idade da sobrevivente feminina mais velha?

4. Qual o percentual de sobreviventes da primeira e terceira classes?

```python
# seu código
sfirst = titanic.loc[(titanic['survived']==1) & (titanic['class']=='First')]
sthird = titanic.loc[(titanic['survived']==1) & (titanic['class']=='Third')]

pric= titanic.loc[titanic['class']=='First']
terc= titanic.loc[titanic['class']=='Third']

porc1= len(sfirst) / len(pric)
porc3= len(sthird) / len(terc)
print(int(porc1*100))
print(int(porc3*100))

```

5. Analise a média de idade dos passageiros por classe e sexo (sugestão: empregue groupby). Você pode notar que?

```python
# seu código

classe_m = titanic.groupby('class').median()
print (classe_m)

sexo = titanic.groupby('sex').median()
print (sexo)
```
